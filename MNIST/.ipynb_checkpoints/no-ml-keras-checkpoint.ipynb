```python
from keras.datasets import mnist
import numpy.typing as npt
import numpy as np
from collections import defaultdict
import seaborn as sns
import cv2
```

```python
from matplotlib import pyplot as plt
# load dataset
(trainX, trainy), (testX, testy) = mnist.load_data("./files")
# summarize loaded dataset
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
# plot first few images
for i in range(9):
 # define subplot
 plt.subplot(330 + 1 + i)
 # plot raw pixel data
 plt.imshow(trainX[i], cmap=plt.get_cmap('gray'))
# show the figure
plt.show()
```

## Thinning:
1. Convert to binary images: 0 -> black, 1-> white
2. repeatedly erode and dilate

```python
binaryTrainX = np.where(trainX > 0, 1, 0)
binaryTrainX = binaryTrainX.astype(np.uint8) # converting to unsigned int is important for open cv methods
```

```python
plt.imshow(binaryTrainX[0], cmap=plt.get_cmap('gray'))
```

```python
kernel = np.ones((3,3),np.uint8) # 3x3 kernel keeps a 1 if all the adjacent cells have 1
erosion = cv2.erode(binaryTrainX[0],kernel,iterations = 1)
print(erosion)
print(binaryTrainX[0])
plt.imshow(erosion, cmap=plt.get_cmap('gray'))
```

```python
def countWhites(img:  npt.NDArray, threshold=200):
    return np.sum(img > threshold)

def countNonZeros(img:  npt.NDArray):
    return np.sum(img > 0)
```

```python
whiteCounts = []
for sample in trainX:
    whiteCounts.append(countWhites(sample, 150))
    
```

```python
whiteCounts[0]
```

```python
trainy[0]
```

```python
whiteDistributions = defaultdict(lambda : [])
for idx, digit in enumerate(trainy):
    whiteDistributions[digit].append(whiteCounts[idx])
```

```python
for k, vals in whiteDistributions.items():
    g = sns.displot(vals)
    g.set_titles(k)
```
