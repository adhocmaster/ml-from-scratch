```python
import cv2
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((512, 512, 3), dtype=np.int16)
```

```python
blank.shape
```

```python
plt.imshow(blank)
```

```python
cv2.rectangle(blank, (250, 0), (350, 100), color=(0, 255, 0), thickness=5)
plt.imshow(blank)
cv2.rectangle(blank, (250, 200), (350, 300), color=(0, 255, 0), thickness=-1)
plt.imshow(blank)
```

```python
cv2.circle(blank, center=(50, 400), radius=10, color=(255, 200, 0), thickness=5)
plt.imshow(blank)
```

```python
cv2.circle(blank, center=(100, 400), radius=10, color=(255, 200, 0), thickness=-1)
plt.imshow(blank)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank, text="hello world", org=(10, 500), fontFace = font, fontScale=2, color=(255, 255, 255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank)
```

```python
poly = np.array([[100, 300], [200, 100], [200, 300], [400, 400]],dtype=np.int32)
```

```python
polyCV2Format = poly.reshape((-1, 1, 2)) # cv2 wants points for each color channel. we have one channel for each point.
```

```python
polyCV2Format
```

```python
cv2.polylines(blank, [polyCV2Format], isClosed=True, color=(0, 0, 255), thickness=2)
plt.imshow(blank)
```
